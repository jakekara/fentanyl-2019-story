```python
import pandas as pd
```

```python
df = pd.read_csv("data/pop/PEP_2017_PEPASR6H/PEP_2017_PEPASR6H_with_ann.csv",skiprows=1)
df.head()
```

```python
df["Id.1"].value_counts()
```

```python
hisp = df[df["Id.1"] == "hisp"]
nhisp = df[df["Id.1"] == "nhisp"]
tot = df[df["Id.1"] == "totpop"]
```

```python
df = tot.set_index("Year")[["Total; Both Sexes; Total"]]
df.columns = ["total"]
df
```

```python
hisp_df = hisp.set_index("Year")[["Total; Both Sexes; Total"]]
hisp_df.columns=["hispanic"]
df = df.join(hisp_df)
df
```

```python
nhisp_df = nhisp.set_index("Year")[["Race Alone - White; Both Sexes; Total",
                        "Race Alone - Black or African American; Both Sexes; Total"]]
nhisp_df.columns = ["white", "black"]
df = df.join(nhisp_df)
df
```

```python
# Fix the years 
df = df[df.index.map(lambda x: "July" in x)]
df.index = df.index.map(lambda x: x.replace("July 1, ",""))
df.to_csv("./data/popest-race.csv")
```

```python
df = df.transpose()
df["2018"] = df["2017"]
df.to_csv("./data/popest-race.csv")
df
```
